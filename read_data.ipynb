```python
import h5py
import pandas as pd
```

```python
path = 'output data/heart_32_main/batch_results.hdf'
```

```python
dataset = {}
with h5py.File(path, 'r') as f:
    print(f['Interaction Data']['Detector at angles starting at 0.0 deg'].keys())
    print(f['Interaction Data']['Detector at angles starting at 0.0 deg']['0.0'].keys())
    for key in f['Interaction Data']['Detector at angles starting at 0.0 deg']['0.0']:
        data = f['Interaction Data']['Detector at angles starting at 0.0 deg']['0.0'][key]
        if key != "process_name":
            dataset[key] = list(data)
df1 = pd.DataFrame(dataset)
```

```python
dataset = {}
with h5py.File(path, 'r') as f:
    print(f['Interaction Data']['Detector at angles starting at 0.0 deg'].keys())
    print(f['Interaction Data']['Detector at angles starting at 0.0 deg']['3.141592653589793'].keys())
    for key in f['Interaction Data']['Detector at angles starting at 0.0 deg']['3.141592653589793']:
        data = f['Interaction Data']['Detector at angles starting at 0.0 deg']['3.141592653589793'][key]
        if key != "process_name":
            dataset[key] = list(data)
df2 = pd.DataFrame(dataset)
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter3d(
    x=df1['global_position'].apply(lambda x: x[0]),
    y=df1['global_position'].apply(lambda x: x[1]),
    z=df1['global_position'].apply(lambda x: x[2]),
    mode='markers',
    marker=dict(
        size=1,
        color='blue'
    ),
    name='Torch 1'
))

fig.add_trace(go.Scatter3d(
    x=df2['global_position'].apply(lambda x: x[0]),
    y=df2['global_position'].apply(lambda x: x[1]),
    z=df2['global_position'].apply(lambda x: x[2]),
    mode='markers',
    marker=dict(
        size=1,
        color='blue'
    ),
    name='Torch 2'
))

fig.add_trace(go.Scatter3d(
    x=df2['emission_position'].apply(lambda x: x[0]),
    y=df2['emission_position'].apply(lambda x: x[1]),
    z=df2['emission_position'].apply(lambda x: x[2]),
    mode='markers',
    marker=dict(
        size=1,
        color='red'
    ),
    name='Emission'
))


fig.show()
```
